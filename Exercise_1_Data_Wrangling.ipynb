## Data Wrangling

#### Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
 "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
 "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
 "peak-rpm","city-mpg","highway-mpg","price"]
```

#### Reading dataset giving header

```python
df=pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv",names=headers)
```

```python
df.head()
```

#### Checking Na values if any

```python
df.isnull().sum()
```

#### Replacing '?' with NaN values

```python
df.replace("?",np.nan,inplace=True)
```

```python
missing=df.isnull()
missing.head()
```

#### Checking missing values in every column

```python
for col in missing.columns.values.tolist():
    print (col)
    print(missing[col].value_counts())
```

```python
df.info()
```

#### Changing the dtype of required columns

```python
df['normalized-losses']=df['normalized-losses'].apply(pd.to_numeric)
df['bore']=df['bore'].apply(pd.to_numeric)
df['stroke']=df['stroke'].apply(pd.to_numeric)
df['horsepower']=df['horsepower'].apply(pd.to_numeric)
df['peak-rpm']=df['peak-rpm'].apply(pd.to_numeric)
```

```python
df.info()
```

#### Delete target column rows that have missing data

```python
df['price']=df['price'].dropna()
```

#### Replacing missing values in the columns with mean of that column

```python
df['normalized-losses']=df['normalized-losses'].fillna(df['normalized-losses'].mean())
df['bore']=df['bore'].fillna(df['bore'].mean())
df['stroke']=df['stroke'].fillna(df['stroke'].mean())
df['horsepower']=df['horsepower'].fillna(df['horsepower'].mean())
df['peak-rpm']=df['peak-rpm'].fillna(df['peak-rpm'].mean())
df['num-of-doors']=df['num-of-doors'].fillna(df['num-of-doors'].mode())
```

#### Resetting index

```python
df.reset_index(inplace=True)
```

```python
df.drop(['index'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['price']=df['price'].apply(pd.to_numeric)
df.info()
```

```python
df
```

#### Transforming city-mpg and highway-mpg into liters/100km 

```python
df['city-L/100km']=235/df['city-mpg']
df['highway-L/100km']=235/df['highway-mpg']
```

#### Normalizing the columns

```python
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

```python
import matplotlib.pyplot as plt
```

#### Distribution of horsepower

```python
plt.hist(df['horsepower'])
```

#### Creating three equal sized binned column horsepower-binned with values -  “low”, “medium”, “high” 

```python
binwidth = (df['horsepower'].max() - df['horsepower'].min())/4 
bins = np.arange(df['horsepower'].min(), df['horsepower'].max(), binwidth)
group_names = ['low', 'medium', 'high'] 
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels = group_names, include_lowest = True)
```

#### Checking distribution

```python
df['horsepower-binned'].hist()
```
