#### Calculating the pearson correlation between “wheel-base” and “price”. 

```python
from scipy.stats import pearsonr
pearsonr(df['wheel-base'],df['price'])
```

Correlation is positive here & the p-value indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.
So here p-values for the correlation is too less than the significance level of 0.05, which indicates that the correlation coefficients are significant. 

#### Performing one way ANOVA test to check if different groups of “drive-wheels” are correlated with “price"

```python
df_new=df[['drive-wheels','price']]
df_new.head()
```

```python
rwd=pd.DataFrame(df_new[df_new['drive-wheels']=='rwd'].values)
rwd.columns=['rwd','price']
rwd_price=list(rwd['price'])
```

```python
fwd=pd.DataFrame(df_new[df_new['drive-wheels']=='fwd'].values)
fwd.columns=['fwd','price']
fwd_price=list(fwd['price'])
```

```python
wd=pd.DataFrame(df_new[df_new['drive-wheels']=='4wd'].values)
wd.columns=['wd','price']
wd_price=list(wd['price'])
```

```python
from scipy.stats import f_oneway
f_oneway(rwd_price,fwd_price,wd_price)
```

One-way ANOVA tests the null hypothesis that two or more groups have the same population mean & p-value tells the differences between the means are statistically significant or not. So here it is too less so we can reject null hypothesis and say that groups have different means
